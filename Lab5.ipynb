```python
from math import sqrt,fabs
my_list=[]
for i in range(22):
    my_list.append(randint(-30,30))
for i in my_list:
    print(i,end=" ")
print()
s=0
neg=0
p=1
for i in my_list:
    if i%2==0:
        s+=i
    if i<0:
        neg+=1
    if fabs(i)<7:
        p*=i
print("s=",s,sep="")
print("neg=",neg,sep="")
print("p=",p,sep="")
```

```python
from random import randint
task=set()
variants=dict()
for i in range(1,23):
    while True:
        n=randint(1,31)
        if n not in task:
            task.add(n)
            variants[i]=n
            break
for key,value in variants.items():
    print(f"Номер {key}: Вариант {value}")
```
