```python
def printBin(n):
    k=128
    while k>0:
        print(n//k,end="")
        n%=k
        k//=2
```

```python
printBin(254)
```

```python
def Line(n):
    for i in range(n):
        print('_',sep="",end="")
```

```python
n=int(input("Введите число:"))
Line(n)
```

```python
def Numbers(n):
    res=""
    while n>0:
        m=n%10
        res+=str(m)
        n//=10
    return "".join(reversed(res))
```

```python
n=int(input("Введите число:"))
Numbers(n)
```
